```python
#Import Dependencies
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file = "Exchange_Rate_Report1.csv"
foreign_ex = pd.read_csv(file)
foreign_ex
```

```python
foreign_ex.columns
```

```python
foreign_ex = foreign_ex.rename(columns={'Unnamed: 1': 'Australian Dollar (AUD)', 'Unnamed: 2': 'Chinese Yuan (CNY)', 'Unnamed: 3': 'Euro (EUR)', 'Unnamed: 4': 'Japanese Yen (JPY)', 'Unnamed: 5': 'Swiss Franc (CHF)', 'Unnamed: 6':'U.K. Pound (GBP)', 'Unnamed: 7': 'U.S. Dollar (USD)'})
foreign_ex
```

```python
foreign_ex.columns
```

```python
foreign_ex_1 = foreign_ex.drop([0, 1])
foreign_ex_1
```

```python
clean_foreign_ex = foreign_ex_1.drop(foreign_ex_1.index[1965:1984])
clean_foreign_ex
```

```python
clean_foreign_ex
```

```python
clean_foreign_ex = clean_foreign_ex.reset_index()
```

```python
clean_foreign_ex.columns
```

```python
clean_foreign_ex = clean_foreign_ex[['Representative rates for the period  April 28, 2013 - March 23, 2021',
       'Australian Dollar (AUD)', 'Chinese Yuan (CNY)', 'Euro (EUR)',
       'Japanese Yen (JPY)', 'Swiss Franc (CHF)', 'U.K. Pound (GBP)',
       'U.S. Dollar (USD)']]
clean_foreign_ex
```

```python
clean_foreign_ex.dtypes
```

```python
clean_foreign_ex['Australian Dollar (AUD)'] = clean_foreign_ex['Australian Dollar (AUD)'].astype(float)
```

```python
clean_foreign_ex['Chinese Yuan (CNY)'] = clean_foreign_ex['Chinese Yuan (CNY)'].astype(float)
```

```python
clean_foreign_ex['Euro (EUR)'] = clean_foreign_ex['Euro (EUR)'].astype(float)
```

```python
clean_foreign_ex['Japanese Yen (JPY)'] = clean_foreign_ex['Japanese Yen (JPY)'].astype(float)
```

```python
clean_foreign_ex['Swiss Franc (CHF)'] = clean_foreign_ex['Swiss Franc (CHF)'].astype(float)
```

```python
clean_foreign_ex['U.K. Pound (GBP)'] = clean_foreign_ex['U.K. Pound (GBP)'].astype(float)
```

```python
clean_foreign_ex['U.S. Dollar (USD)'] = clean_foreign_ex['U.S. Dollar (USD)'].astype(float)
```

```python
#Mean for AUD, CNY, EUR, JPY, CHF, GBP

AUD_mean = clean_foreign_ex['Australian Dollar (AUD)'].mean()
print(f"The mean for AUD is {AUD_mean}")
```

```python
CNY_mean = clean_foreign_ex['Chinese Yuan (CNY)'].mean()
print(f"The mean for CNY is {CNY_mean}")
```

```python
EURO_mean = clean_foreign_ex['Euro (EUR)'].mean()
print(f"The mean for EURO is {EURO_mean}")
```

```python
JPY_mean = clean_foreign_ex['Japanese Yen (JPY)'].mean()
print(f"The mean for JPY is {JPY_mean}")
```

```python
CHF_mean = clean_foreign_ex['Swiss Franc (CHF)'].mean()
print(f"The mean for CHF is {CHF_mean}")
```

```python
GBP_mean = clean_foreign_ex['U.K. Pound (GBP)'].mean()
print(f"The mean for GBP is {GBP_mean}")
```

```python
USD_mean = clean_foreign_ex['U.S. Dollar (USD)'].mean()
print(f"The mean for USD is {USD_mean}")
```

```python
cleaned_foreign_ex = clean_foreign_ex.fillna(method='ffill')
cleaned_foreign_ex
```

```python
# Push the remade DataFrame to a new CSV file
cleaned_foreign_ex.to_csv("cleaned_foreign_ex.csv",
                  encoding="utf-8", index=False, header=True)
```

```python
#Fluctuations
cleaned_foreign_ex.plot(x="Representative rates for the period April 28, 2013 - March 23, 2021",y="Australian Dollar (AUD)", rot=45)
plt.xlabel("Date",size=16)
plt.ylabel("AUD",size=16)
plt.title("AUD Fluctuations", size=18)
plt.tight_layout()
```
